# Remove mitochondrial genes & human genes

```python
mtm = pd.Series(df['transcript']).str.contains('mt-').values # startswith() is case sensitive
mth = pd.Series(df['transcript']).str.contains('MT-').values
human = pd.Series(df['transcript']).str.isupper()
```

```python
x = copy.deepcopy(df)
```

```python
matches = [i for i in range(0,len(human)) if human[i] == True]
```

```python
len(matches)
```

```python
x = x.drop(matches);
```

# Remove ribosomal genes

```python
mRb = pd.Series(x['transcript']).str.contains('Rb').values
mRp = pd.Series(x['transcript']).str.contains('Rp').values
matches = [i for i in range(0,len(mRb)) if mRb[i] == True or mRp[i] == True or mtm[i] == True]
x = x.drop(matches);
```
